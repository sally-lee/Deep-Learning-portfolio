##Weight initialization for deep networks

- - -

###학습목표

- - -

아주 깊은 신경망에서 경사 소실과 폭발의 문제에 도움을 줄수 있는 해결방법을 알아보자.

###Single neuron example

- - -

101.jpg

　　z=$w_{1}$$x_{1}$+$w_{2}$$x_{2}$ $\cdots$+ $w_{n}$$x_{n}$ 

* b=0 이라고 하고 지금은 b를 무시한다.
따라서 z의 값이 너무 크거나 작아지지 않도록 만들어야한다.

* n값이 클수록  $w_{i}$값이 작아져야한다. 왜냐하면 z는  $w_{i}$ $x_{i}$의 합이기 때문에 이 많은 항들을 더하면 각각의 항이 작아지기를 바랄것이기 때문이다


* Var($w_{i}$)=$\frac{1}{n}$ 여기서 n은 뉴런으로 들어가는 입력 특성의 개수이다.


* 실제로 특정 층에 대한 가중치 행렬 $w^{[l]}$=np.random.randn(shape)$\times$ np.sqrt($\frac{1}{n^{[l-1]}}$) 이다.

 $\rightarrow$ 만약 ReLU 함수를 사용한다면 $w^{[l]}$=np.random.randn(shape)$\times$ np.sqrt($\frac{2}{n^{[l-1]}}$) 이다.


*  $g^{[l]}$(z)=ReLU(z) 

* $n^{[l-1]}$울 사용하는 이유는 이예제의 로지스틱 회귀에서는 n개의 입력 특성을 갖지만 더 일반적인 경우에는 층ㅣ은 해당 층의 각 유닛에 대해 $n^{[l-1]}$의 입력을 갖는다.

*  $w^{[l]}$=np.random.randn(shape)$\times$ np.sqrt($\frac{2}{n^{[l-1]}}$) 에서 $\times$ np.sqrt($\frac{2}{n^{[l-1]}}$) 해주는 이유는 몰까?
 
 $\rightarrow$가중치 행렬 w를 1보다 너무 커지거나 너무 작아지지 않게 설정해서 너무 빨리 폭발하거나 소실되지 않게 하기 위해서이다.


###다른 변형예제(1)

- - -

ReLU활성화 함수 대신 tanh활성화 함수를 사용하다면?

 $\rightarrow$ $w^{[l]}$=np.random.randn(shape)$\times$ np.sqrt($\frac{1}{n^{[l-1]}}$)를 해준다. 여기서 np.sqrt($\frac{1}{n^{[l-1]}}$)를 Xavier Initialization라고 한다.


###다른 변형예제(2)

- - -

  $w^{[l]}$=np.random.randn(shape)$\times$np.sqrt($\frac{2}{n^{[l-1]}+n^{[l]}}$)을 쓴다.

###Weight initialization의 목표 

- - -

* 경사 소실과 폭발 문제에 대한 직관을 얻는데 도움이 된다.

* 가중치를 초기화하는 합당한 크기를 선택하는데 도움이 된다.

* 가중치가 너무 빨리 폭발하거나 0으로 수렴하지 않게 도와준다.따라서 가중치 혹은 경사의 폭발과 소실 없이 상당히 깊은 네트워크를 훈련시킬수 있다.
