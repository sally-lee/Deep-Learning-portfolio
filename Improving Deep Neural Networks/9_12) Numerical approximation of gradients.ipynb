###왜 더 큰 삼각형에서  $\frac{높이}{밑변}$를 구하는것이 θ에서의 도함수를 근사하는데 더 나은 값을 제공하는가.

- - -



* $\frac{f(θ+ε)-f(θ-ε)}{2ε}$ $\approx$ g(θ)

* $\frac{(1.03)^{3}-(0.99)^{3}}{2(0.01)}$=3.0001

1. 이전슬라이드에서 g(θ)=$3θ^{2}$였고 θ가1일때 g(θ)의 값은 3이였다.3.0001와 3은 매우 가깝다.근사 오차는 0.0001이다.


2. 이전슬라이드에서 g(θ)=$3θ^{2}$였고 θ가 θ+ε인 1.01일때 g(θ)의 값은 3.0301이였다.근사오차는 0.03이다.




 $\rightarrow$ 따라서! 도함수를 근사하기 위 **1**.처럼 양쪽 차이를 이용하는 방법을 사용하면  g(θ)가 f의 도함수에 대한 더 올바른 구현이라는 것을 알수있다.



 




### 미적분을 이용한 추가적인 이론

- - -

$$\lim_{x \to \ 0} \frac{f(θ+ε)-f(θ-ε)}{2ε}={f}'(θ)$$




* ε이 0 이 아닌 값에 대해서 이 근사의 오차는 o($ε^2$)이다.

* ε는 매우 작은 숫자이다.

* ε이 0.01인 경우 $ε^2$ 은 0.0001이된다.

* o($ε^2$)와 같이 표기하는 Big-O (빅 오 표기법)은 오차가 상수라는 것을 나타낸다.

- - -

$$\lim_{x \to \ 0} \frac{f(θ+ε)-f(θ)}{ε}={f}'(θ)$$

* 오차는 o(ε) 이다.

* ε이 1보다 작은 값이므로 $ε^2$보다는 훨씬 큰 값이다.
