여기 적은 수의 층을 가진 신경망이 있습니다 

수정1.jpg

하나의 층을 골라서 그 층에 집중하여 계산을 해봅시다

a. layer $l$에 대해서 파라미터 $W^{[l]}$과 $b^{[l]}$이 있습니다  
 * layer $l: W^{[l]},b^{[l]}$  


b. 정방향 전파에 대해서는 입력으로
이전 층의 활성값인 $a^{[l-1]}$을 주게 됩니다 그럼 출력은 $a^{[l]}$이 됩니다  
* $Forward$ : $input$ $a^{[l-1]}$, $output$ $a^{[l]}$ 


c.  $z^{[l]}$은 $W^{[l]}×a^{[l-1]}+b^{[l]}$과 같습니다 그럼 $a^{[l]}$은 $g^{[l]}(z^{[l]})$과 같습니다
* $z^{[l]} = W^{[l]}a^{[l-1]}+b^{[l]}$  
$a^{[l]} = g^{[l]}(z^{[l]})$   
  


* $z^{[l]}$의 값을 저장하면 역전파의 단계에서 유용하게 쓸 수 있습니다.
 * $cache$ $z^{[l]}$ 
   


d. 역전파의 단계에서 다시 layer $l$에 대한 계산을 하면
입력으로 $da^{[l]}$을 받고
출력으로 $da^{[l-1]}$이 나옵니다
* $Backward$ : $input$ $da^{[l]}$, $output$ $da^{[l-1]}$  
입력은 $da^{[l]}$과
저장해 놓은 $z^{[l]}$을 사용합니다


e. $da^{[l-1]}$과 경사 하강법 학습을 위한 그래디언트를 출력합니다
* $dW^{[l]}$, $db^{[l]}$



$\Rightarrow$  이것이 정방향과 역방향 단계를 구현하기 위한 기본 구조이며 정방향 함수와 역방향 함수라고도 할 수 있다

b. 첫 번째 층 활성인 $a^{[1]}$를 계산합니다
 * 이것을 위해 $W^{[1]}$과 $b^{[1]}$를 사용합니다
 *  $z^{[1]}$을 $cache$에 저장합니다

c. $a^{[1]}$ 을 두 번째 층에 집어넣습니다.


d. 다음 층의 활성인 $a^{[2]}$를 계산합니다.
 * $W^{[2]}$와 $b^{[2]}$를 사용합니다
 * $z^{[2]}$을 $cache$에 저장합니다

e. 이런 식으로 반복합니다

- 이 값은 $y$의 예측값과 같습니다

$\blacksquare$  모든 층의 $z$ 값은 $cache$에 저장합니다

 -  $cache$을 실제로 구현할 때 파라미터 $W^{[l]}$과 $b^{[l]}$도 저장하는게 편리하다

## 2)역방향 전파 단계
역전파 단계에서는 정방향 전파을 역순으로 하는 것입니다


수정5.PNG

a. $da^{[l]}$을 가져와서 집어넣습니다


b. $da^{[l-1]}$을 출력합니다



c. 이런식으로 반복하여 $da^{[1]}$ 까지 계산합니다.
 * $da^{[0]}$를 계산하여 하나의 출력을 더 얻을 수 있습니다 그러나 이것은 입력 특성에 대응하는 도함수이고
최소한 지도 신경망의 가중치를 학습하는 것에는 유용하지 않습니다
 * 역전파의 과정에서 $dW^{[l]}$과 $db^{[l]}$을 출력합니다
즉 필요한 도함수는 모두 계산하게 됩니다

a. 신경망의 하나의 학습 반복은 $x$값인 $a^{[0]}$로 시작해서 정방향 전파를 따라서 $y$의 예측값을 계산하고 이 값으로 $da^{[l]}$을 계산하고
역전파를 거쳐서 모든 도함수를 얻게 됩니다



b. $W^{[l]}$의 값은 학습률을 곱해 $W^{[l]}-$ $\alpha$$dW^{[l]}$가 됩니다  
$b$에 대해서도 마찬가지로 역전파를 계산해서 모든 도함수 값을 알 수 있습니다
  * $W^{[l]}$:= $W^{[l]}-$ $\alpha$$dW^{[l]}$ ($\alpha$ = 학습률)
  * $b^{[l]}$:= $b^{[l]}$ $-$ $\alpha$$dW^{[l]}$
