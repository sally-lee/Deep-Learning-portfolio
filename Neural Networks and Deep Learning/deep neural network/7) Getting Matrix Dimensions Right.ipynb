## forward propagation 구현

### W,b,dW,db차원인 경우

zwxb차원.png



1.   $z^{[1]} = W^{[1]} \times x(입력특성) + b^{[1]}$
     
     이때 편향 값 $b^{[1]}$은 무시해도 된다
*   파란색원은 첫번째 은닉층을 표시한 것으로 3개의 히든 유닛이 있다.
*  층은 0,1,2,3,4,5로 이루어져 있고, 층 1의 히든 유닛의 개수는 3개로 $n^{[1]} = 3$으로 표시하고, 다른 층 또한 $n^{[2]} = 5$, $n^{[3]} = 4$, $n^{[4]} = 3$, $n^{[4]} = 2$, $n^{[5]} = 1$로 표현한다
*   입력층은 $n^{[0]} = n<sub>x</sub> = 2로 표현한다

2.  z, W, x의 차원을 생각하기
*   z는 첫번째 은닉층에 대한 활성화 벡터로 3차원 벡터로 (3,1)행렬이다
*   x에는 두가지 입력 특성이 있어 (2,1) 행렬로 $(n^{[0]},1)$이다
*   $W^{[1]}$은 $(n^{[0]},1)$벡터를 곱했을 때 $(n^{[1]},1)$벡터가 나오도록 하는 값이어야 한다
*   따라서 이런 식의 3차원 벡터는 무언가에 2차원 벡터를 곱한 것과 같고, 행렬 곱의 규칙에 따라 (3,2)행렬이 됨으로써 (3,2)행렬 곱하기 (2,1)행렬은 (3,1) 벡터가 되기 때문이다

3. $W^{[1]}$이 $(n^{[1]},n^{[0]})$가 되어야 하는 이유
*   (3,2)행렬 곱하기 (2,1)행렬은 (3,1)이 됨으로 써 $(n^{[1]},n^{[0]})$차원의 행렬이 된다
 
 즉 $w^{[l]}$이 $(n^{[l]},n^{[0]})$ 라는 것이다

  <font color="green"> 예시














*   $W^{[2]}$의 차원은 해당 층의 차원과 그 전 층의 차원(5,3)으로 $(n^{[2]}$,$n^{[1]})$과 같다
*   $z^{[2]}$를 계산하면 $W^{[2]}×a^{[1]}$이고 z는 (5,3), $a^{[1]}$는 (3,1), $W^{2]}$는 (5,3)의 행렬을 갖는다 

*   $W^{[2]}$와 같은 원리로 $W^{[3]}$: (4,5), $W^{[4]}$ :(2,4), $W^{[5]}$ : (1,2) 가 되고 $W^$의 차원은 $W^{[l]}$이 $(n^{[l]},n^{[l-1]})$ 정의 할 수 있다


4. 차원


*   $W^{[1]} \times x + b^{[1]}$인 경우
  
   $W^{[1]} \times x$의 벡터가 (3,1) 되고, 출력으로 (3,1)벡터를 얻기 위해 또 다른(3,1)벡터를 더해줘야된다 
   그래서 $b^{[1]}$은 $(n^{[1]},1)$이고 값은 (3,1)이 된다
*    $W^{[2]}×a^{[1]} + b^{[2]}$인 경우

     $W^{[2]}×a^{[1]}$의 벡터가 (5,1)되고, $W^{[2]}×a^{[1]} + b^{[2]}$가 (5,1)벡터가 되기 위해 또 다른 (5,1) 벡터의 값이 필요하다
     그래서 $b^{[2]}$는 $(n^{[2]},1)$이고 값은 (5,1)이 된다

*   따라서 $b^{[l]}$는 $(n^{[l]},n^{[l-1]})$의 차원이 되어야 된다



## backward propagation 구현

역전파 구현.png

*   $dW^{[l]}$의 차원은 $W^{[l]}$의 차원과 같아야 되고, $db^{[l]}은 b^{[l]}$과 같은 차원을 가져야한다
*   차원을 확일 할 때 중요한 값은 z,x,$a^{[l]}$인데 $z^{[l]}$이 $g^{[l]}(a^{[l]})$과 같기 때문에 이런 종류의 네트워크에서 z와 a의 차원은 같아야 한다
*   벡터화 W,b,dW,db의 차원도 같아진다





### z, a ,x의 차원인 경우

z , a , x 차원.png



1.   $z^{[1]}$ = $W^{[1]}\times x + b^{[1]}$
*   $z^{[1]}$는 $(n^{[1]},1)$, $W^{[1]}$는 $(n^{[1]},n^{[0]})$, x는 $(n^{[1]},0)$,$b^{[1]}$는 $(n^{[1]},1)$로 각각의 행렬을 갖는다

2.   $Z^{[1]}$ = $W^{[1]}\times X + b^{[1]}$
*  $Z^{[1]}$ 는  $z^{[1]}$, $z^{[2]}$,...,$z^{[m]}$까지의 데이터값을 저장하고, $(n^{[1]},1)$이 아닌 $(n^{[1]},m)$의 차원을 갖게 된다
*   $W^{[1]}$는 $(n^{[1]},n^{[0]})$차원으로 그대로이고,  X는 $(n^{[1]},0)$가 아닌 모든학습 데이터가 수평으로 저장되기 때문에 $(n^{[0]},m)$이 된다
*  $W^{[1]}$행렬 $(n^{[1]},n^{[0]})$를 X행렬$(n^{[0]},m)$에 곱하면 $(n^{[1]},m)$이 나온다
*   $b^{[1]}$도 여전히 $(n^{[1]},1)$이였지만, b에 더한다면 브로드캐스팅을 통해 $(n^{[1]},m)$행렬이 되고 원소별로 더해지게 된다

3. $z^{[1]}, a^{[1]} 은 (n^{[l]},1)$ 차원을 가지고 $Z^{[l]}, A^{[1]} 은 (n^{[l]},m)$ 차원을 갖는다
*   l = 0인 xmr별한 경우에서 $A^{[0]}$가 훈련 집합의 입력특성 X와 같고, 값은 $(n^{[0]},m)$가 된다
* 역전파를 구현할 때

  $dZ^{[l]}$와 $dA^{[0]}$도 계산을 하는데 $dZ^{[l]}$와 $dA^{[0]}$는 Z,A와 같은 차원을 가진다






심층 신경망의 backforword propagation할 때 코드를 쭉 보면서 모든 행렬의 차원이 일치하는지 잘 확인하고, 나올만한 일부 버그를 삭제하는데 도움이 된다.
