###**음성의 시각화 주요원리**

- - -

1.  여러 개의 은닉층을 갖는 심층 신경망은 낮은 단계에서는 간단한 특징을 학습하게 되고 그 후 깊은 층에서는 탐지된 간단한 것들을 함께 모아
단어, 구, 문장 같은 더 복잡한 것들을 탐지하게 된다.

2. 초기의 층에서 계산되는 것은 상대적으로 간단한 함수이다.
ex) 모서리의 위치를 찾는 것

3. 네트워크가 더 깊어질수록 놀라울만큼 복잡한 것들이 가능해진다.
ex)얼굴을 인식하거나 단어, 구, 문장을 인식하는 것

4. 어떤 사람들은 심층 신경망을 사람의 뇌에 비유하기도 한다.

* 왜 심층 신경망을 사람의 뇌에 비유할까?  $\rightarrow$ 신경 과학자가 말하듯이 사람의 뇌도 심층신경망처럼 모서리같이 눈에 보이는 간단한 것부터 감지하고 이를 모아 더 복잡한 정보를 인식하기때문이다.
