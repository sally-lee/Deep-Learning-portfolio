# **Binary Classification**


## **1. Binary Classification이란?**
---
* 입력값에 따라 모델이 분류한 카테고리가 두 가지인 분류 알고리즘
* 어떤 대상에 대한 규칙이 1(True)인지 0(False)인지를 분류하는데 쓰인다.

## **2. Data preprocessing**이란?
----







* 기존의 데이터를 머신러닝 알고리즘에 알맞은 데이터로 바꾸는 과정이다.


### 예시


수정중.jpg

이미지가 입력값으로 주어졌을 때
고양이 이미지라면 $1$을 출력하고  
고양이 이미지가 아니라고 판단하면 $0$을 출력하려한다.  

**Training data를 모델의 input data로 변환**

1.이미지의 빨간 초록 파란색 채널들과 일치하는 세 개의 행렬을 따로따로 저장.
 * 입력된 이미지이 64 x 64픽셀이라면 세 개의 64 x 64 행렬들이 이미지의 빨간 초록 파란색 픽셀 강도 값을 나타낸다.  

2.픽셀 강도 값들을 feature  벡터로 바꾸기 위해서
모든 픽셀값을 feature  벡터 x의 한 열로 나열. 
 * 픽셀값인 빨간 초록 파란색 픽셀 강도 값들을 전부를 차례대로 나열.
 * 픽셀값을 하나의 벡터로 펼치기 위해서 주어진 이미지에 대한 feature 벡터 x를 정의한다.
$$x=\begin{bmatrix}
\\ 1
\\ 2
\\ .
\\ .
\\ .
\\ m
\end{bmatrix} $$



* 이미지가
64 x 64일 경우 벡터 $x$의
전체 차원은 64 x 64 x 3가 된다.  
$ 64 * 64 * 3 = 12288$
* 12,288인 입력 feature  벡터 $x$의
차원을 $n_{x}$로 표현한다.  
$n^{x} = 12288 $


##**3. Binary Classification 의 목표**
---
입력된 이미지를 나타내는 feature 벡터 $x$를 가지고

그에 대한 레이블 $y$ $1$ 아니면 $0$

다시 말해 고양이 이미지인지 아닌지를

예측할 수 있는 분류기를 학습하는 것이다.

##**4. 자주쓰는 표기법**
---

훈련 샘플을 한쌍 $(x,y)$ 로 표시할때

훈련 세트는 $m$개의 훈련 샘플을 포함하고 다음과 같이 표기될 것이다.

$m$ traing example : {$(x^{(1)},y^{(1)}),(x^{(2)},y^{(2)}),\ldots (x^{(m)},y^{(m)})$}

모든 훈련 샘플들을 더 간결하게 표기하기 위해 행렬 $X$를 정의

$X=\begin{bmatrix}
 \vdots &  \vdots&  \vdots&  \vdots\\
 x^{(1)}&  x^{(2)}&  \cdots &  x^{(m)}\\
 \vdots&  \vdots&  \vdots&  \vdots\\
\end{bmatrix}$

$m$이 훈련 샘플의 개수일때 $X$는 $m$개의 열과 $n_x$개의 행들로 이루어져 있다.

$ X\in\mathbb{R}^{n_x*m}$   
$X.shpae = (n_x,m)$


신경망 구성을 더 쉽게하기위해 행렬$Y$를 정의

$Y = \left [y^{(1)} y^{(2)}\cdots y^{(m)} \right ]$    
$ Y\in\mathbb{R}^{1*m}$   
$Y.shpae = (1,m)$
