#**Logistic regression**


## **1. Logistic regression**이란?
---


* 데이터가 특정 범주에 속할 확률을 0에서 1 사이의 값으로 예측하고 그 확률에 따라 가능성이 더 높은 범주에 속하는 것으로 분류해주는 지도 학습 알고리즘

###**예시**
* 고양이 이미지인지 아닌지 구분하고 싶은 이미지가 주어졌을 때 $y$의 예측값을 출력하는 알고리즘을 원한다
 * $y$의 예측값은 입력 feature x가 주어졌을 때 $y$가 1일 확률을 뜻한다  
   즉 입력데이터 $x$가 고양이일 확률을 알려주는 예측값을 생성한다라는 뜻.

* Parameter $w$와 $b$는 각각 벡터와 실수입니다  
$ w\in \mathbb{R}^{n_x}$ $b\in \mathbb{R}$
* 이때 $\hat{y}$을 구해야한다

#### **틀린 예시**
$y$의 예측값을
$w$의 전치 * $x$ $+$ $b$로 놓는다 가정한다
$$\hat{y} = w^{T}x+ b$$
$y$의 예측값은 $y$가 1일 확률을 원하기 때문에  
항상 $0$과 $1$사이어야하는데
$w$의 전치 * $x$ + $b$는 1보다 훨씬 크거나
음수일 수도 있어서
$0$과 $1$ 사이어야 하는 확률의 값으로 성립하지 않는다

####**맞는 예시**


Logistic regression 에서는 $y$의 예측값이

Sigmoid function 를 적용하여 출력된 값이다.
$$\hat{y} = \sigma(w^{T}x+ b)$$

## **2. Sigmoid function이란?**
---

슬라이드0003.jpg

### **Sigmoid function의 공식**

$$\sigma (z) = \frac{1}{1+e^{-z}}$$   






Sigmoid는

0부터 1까지 올라가고
수직축은 0.5(중간값)에서 교차합니다








* $z$ 값이 매우 크다면
$ \sigma (z) \approx \frac{1}{1+0} = 1 $

* $z$ 값이 매우 작다면 
$ \sigma (z) \approx \frac{1}{1+\infty} \approx 0 $



## **3. 결론**
---
* logistic regression 를 구현할 때
 $y$가 1일때 예측값이 1로 잘 예측하도록 파라미터 $w$와 $b$를 학습해야 합니다
