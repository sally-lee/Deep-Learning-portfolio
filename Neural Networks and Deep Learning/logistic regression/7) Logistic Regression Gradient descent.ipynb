###2. Logistic regression recap
- - -

$z= w^Tx+b$      

$\hat{y}^\ =a=\sigma(z)$ 

L(a,y)=$-(y$$log(a)+(1-y)$$log(1-a))$

$\hat{y}^\  : y의 예측값 $

L(a,y) : 샘플하나만 생각할때 그 하나의 샘플에 대한 손실 함수

a: 로지스틱 회귀의 출력값

y:정답 레이블


###3. 로지스틱 회귀를 하는목적은 무엇일까?
- - -


파라미터 w와 b를 변경해서 L(a,y)이 손실을 줄이는 것이다.

###4. Backpropagation으로 가서 a에 대한 손실 함수의 도함수 계산을 해보자
- - -



####**1st step**

<손실함수 미분에 필요한 log미분법>

 y=lnx  $\Rightarrow$  $y'= \frac{1}{x}$

<손실함수>

 L(a,y)=$-(y$$log(a)+(1-y)$$log(1-a))$

 <손실함수 미분>

$\frac{dL(a,y)}{da}$ = $\frac{-y}{a} + \frac{1-y}{1-a}$
















####2nd step

4.jpg


 $\frac{da}{dz}$ = a(1-a)     $\rightarrow$
 $\frac{da}{dz}$가  a(1-a)라는것은 시그모이드 함수 미분 유도과정을 보면 알수있다.


####3rd step

2.jpg





$\frac{dL}{dz}$ = $\frac{dL(a,y)}{dz}$    (두표기법 모두 괜찮다.)

=$\frac{dL}{da}$$\times$$\frac{da}{dz}$

=a-y


* 여기서 $\frac{da}{dz}$ = a(1-a)     $\rightarrow$
 $\frac{da}{dz}$가  a(1-a)라는것은 시그모이드 함수 미분 유도과정을 보면 알수있다.
그리고 $\frac{dL(a,y)}{da}$ = $\frac{-y}{a} + \frac{1-y}{1-a}$ 이다.

위에 식을 참고하여 대입하면  $\frac{dL}{da}$$\times$$\frac{da}{dz}$= ($\frac{-y}{a} + \frac{1-y}{1-a}$) $\times$ a(1-a) 이다. 이 식을 정리하면 a-y이다.

### 5. Backpropagation 의 마지막 단계란 무엇인가
- - -

w와 b를 얼마나바꾸어야 하는지 계산하는것이다.

### 6. 단일 샘플에 대한 경사 하강법의 한 단계
- - -

$dw_1=\frac{dL}{dw_1}$=$x_1$dz

$dw_2=\frac{dL}{dw_2}$=$x_2$dz

db=dz

$w_1$ :=$w_1$ - $\alpha$ d$w_1$


$w_2$ :=$w_2$ - $\alpha$ d$w_2$


$b:= $b- $\alpha$ d$b$



### 7. 단일 샘플에 대한 경사 하강법의 한계
- - -

도함수를 계산하고 단일 샘플에 대한 로지스틱 회귀의 경사 하강법을 구현한는 방법을 알아보았다. **하지만 로지스틱 회귀 모델을 훈련시키려면 단일 샘플이 아니라 m개의 훈련 샘플을 가진 훈련 세트 전체를 훈련해야한다.**
