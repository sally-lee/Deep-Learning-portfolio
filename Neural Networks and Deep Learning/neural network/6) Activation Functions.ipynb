# **Activation Functions**

##**1. 개요**
---

신경망을 만들 때 선택해야 해야하는 것 중 하나는

hidden layer 과 input layer 에서 어떤 Activation Function 를 사용할지 정하는것이다.

지금까지는 Sigmoid Activation Function 를 썼지만 다른 Function이 더 좋을 수도 있다. 

$z^{[1]} = W^{[1]}x + b^{[1]}$  

$a^{[1]} = \sigma(z^{[1]})$   $\rightarrow$ $a^{[1]} = g^{[i]}(z^{[1]})$

$z^{[2]} = W^{[2]}a^{[1]} + b^{[1]}$ 

$a^{[2]} = \sigma(z^{[2]})$ $\rightarrow$ $a^{[2]} = g^{[i]}(z^{[2]})$

여기 쓴 것 처럼 다른 Function인 $g(z^{[1]})$를 쓸 수 있다.  
$g$는 Sigmoid Function가 아닌 Nonlinear function 일 수도 있다.

##**2. Activation Functions**
---

###**1) Sigmoid Function , Tach Functions** 

슬라이드0003.jpg

* **Sigmoid Function**
 * $Sigmoid(z) = \frac{1 }{1+e^{-z}}$
 * $0\leq  Sigmoid(z)\leq  1$
 * $y$가 $0$ or $1$이라면 $\hat{y}$ 은 $-1$과 $1$ 사이 대신 $0$과 $1$사이로 출력하는 게 유리하여 Binary Classification 를 할때는 Sigmoid Activation Function 를 쓴다.  
 $\Rightarrow$ Binary Classification 이외에는 거의 쓰지 않는다.
* **Tach Function** 
 * $tanh(z) = \frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$
 * $-1\leq  tanh(z)\leq  1$
 * $y$가 $+1$ 와 $-1$ 사이여서 평균값이 $0$에 더 가깝기 때문에  
은닉 유닛에 대해 $g(z^{[1]})$ 을 $tanh(z)$ 로 놓는다면 대부분의 상황에서 Sigmoid Function보다 우수하다.
  * Sigmoid Function 와 비슷하지만 중간값이$0$ 이라서 다음 층의 학습을 더 쉽게 해준다.

* **Sigmoid Function 와 Tach Function 의 단점**

 * $z$가 굉장히 크거나 작으면 함수의 도함수가 굉장히 작아진다  
$\Rightarrow$ 함수의 기울기가 $0$에 가까워지고 경사 하강법이 느려진다.  
$\Rightarrow$  학습속도가 느려진다.  


### **2) ReLu Function , Leaky ReLu Function**


슬라이드0004.jpg

 * **ReLu Function**
  * $ReLu(z)= max(0,z)$
  * $ReLu(z)= \left\{\begin{matrix}
 z& z\geq 0 \\
0 & else \\
\end{matrix}\right.$
  * 일반적으로 ReLU Activation Function이 기본값으로 많이 사용된다.
  * $z$가 $z>0$일 때는 도함수가 $1$이고 $z$가 $z<0$이면 도함수가 $0$이 된다.  
   $\Rightarrow$
엄밀하게는 $z$가 $0$일 때의 도함수는 정의되지 않았지만 $z$가 $0$일 때 도함수가 $1$이나 $0$이라고 가정해도 작동한다.

* **Leaky ReLu Function**
 * $Leaky ReLu(z)= max(0.01z,z)$
 * $Leaky ReLu(z)= \left\{\begin{matrix}
 z& z\geq 0 \\
z*a & else \\
\end{matrix}\right.$
 * $z<0$ 일 때 도함수가 $0$ 대신 약간의 기울기를 준다

* **ReLu Function 와 Leaky ReLu Function 장점**
 * 대부분의 $z$에 대해 기울기가 $0$과 매우 다르다.  
 $\Rightarrow$
Sigmoid 나 Tach 대신 ReLu Function 를 쓴다면
신경망은 훨씬 더 빠르게 학습 할 수 있다.


* **ReLu Function 단점**
 * $z$가 음수일 때 도함수가 $0$이라는 것  
  $\Rightarrow$  함수의 기울기가 $0$에 가까워지면 학습속도가 느려진다.
