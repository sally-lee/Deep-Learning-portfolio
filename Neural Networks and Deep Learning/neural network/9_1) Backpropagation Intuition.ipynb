###4.Summary of gradient descent
- - -


<단일 훈련 플에서 도출한 Backpropagation 식>
* d$z^{[2]}$ = $a^{[2]}$ - y

* d$w^{[2]}$=d$z^{[2]}$ $a^{[1]T}$

* d$b^{[2]}$=d$z^{[2]}$

* d$z^{[1]}$=$w^{[2]T}$d$z^{[2]}$$\times$${g^{[1]}}'$($z^{[1]}$)

* d$w^{[1]}$=d$z^{[1]}$$x^T$


* d$b^{[1]}$=d$z^{[1]}$



**But 단일 샘플만을 가지고 하는 것보다는 여러 훈련 샘플을 벡터화하여 처리하는 것이 좋다.**


$z^{[1]}$= $W^{[1]}$x + $b^{[1]}$
,$a^{[1]}$ =$g^{[1]}$($z^{[1]}$ ) $\overset{백터화}{\rightarrow}$ $Z^{[1]}$=$W^{[1]}$X + $b^{[1]}$ ,
$A^{[1]}$ =$g^{[1]}$($Z^{[1]}$) 


          
_ _ _


<벡터화된 Backpropagation식>


 * d$Z^{[2]}$ = $A^{[2]}$- Y

* d$W^{[2]}$=$\frac{1}{m}$d$Z^{[2]}$ $A^{[1]T}$

* d$b^{[2]}$=$\frac{1}{m}$np.sum(d$Z^{[2]}$ ,axis=1,keepdims=True)






* d$Z^{[1]}$=$W^{[2]T}$d$Z^{[2]}$$\times$${g^{[1]}}'$($Z^{[1]}$)

  * 단,d$Z^{[1]}$=$W^{[2]T}$d$Z^{[2]}$$\times$${g^{[1]}}'$($Z^{[1]}$)식에 쓰인 **$\times$는 요소별 곱셈을 의미!!**
  * d$Z^{[1]}$의 차원은 ($n^{[1]}$,m)이다.
  * $W^{[2]T}$의 차원은 ($n^{[1]}$,m)이다.
  * ${g^{[1]}}'$($Z^{[1]}$)의 차원은 ($n^{[1]}$,m)이다.
* d$W^{[1]}$=$\frac{1}{m}$d$Z^{[1]}$$X^T$

* d$b^{[1]}$=$\frac{1}{m}$np.sum(d$Z^{[1]}$ ,axis=1,keepdims=True)
