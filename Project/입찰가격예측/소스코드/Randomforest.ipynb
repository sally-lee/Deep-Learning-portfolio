```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import matplotlib as mpl

# 유니코드 깨짐현상 해결
mpl.rcParams['axes.unicode_minus'] = False

# 나눔고딕 폰트 적용
plt.rcParams["font.family"] = 'NanumGothic'

# 경고 무시
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
train =   pd.read_excel('/content/drive/MyDrive/2차입찰금액 예측모델/dataset/2002_2021 최종.xlsx')
```

```python
train.columns
```

```python
from sklearn.model_selection import train_test_split
x = train[['기초금액', '추정가격', '낙차하한율','예가범위']]
y = train[['입찰금액']]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import math
```

#최적의 하이퍼파라미터 찾는 코드

```python
X_train, X_test, y_train, y_test = train_test_split(x, y ,test_size=0.1,random_state=0)

 
 
params = {
    'n_estimators':(100, 200),
    'max_depth' : (5, 8),
    'min_samples_leaf' : (8, 18),
    'min_samples_split' : (8, 16)
}
rf_run = RandomForestRegressor(random_state=0, n_jobs=-1)
grid_cv = GridSearchCV(rf_run, param_grid=params, cv=2, n_jobs=-1)
grid_cv.fit(X_train, y_train)
 
 
print('최적 하이퍼 파라미터:', grid_cv.best_params_)
print('최적 예측 정확도: {0:.4f}'.format(grid_cv.best_score_))
```

#모델 학습

```python
# 랜덤포레스트 득점모델 학습
rf_run = RandomForestRegressor(random_state=0, max_depth=8, min_samples_leaf=8, min_samples_split=8,n_estimators=200)
rf_run.fit(X_train, y_train)
```

#평균제곱근 오차인 rmse를 이용해서 성능 평가를 해보았다.

```python
# train rmse
train_predict = rf_run.predict(X_train)
print("RMSE':{}".format(math.sqrt(mean_squared_error(train_predict, y_train))) )
 
# validation rmse
test_predict = rf_run.predict(X_test)
print("RMSE':{}".format(math.sqrt(mean_squared_error(test_predict, y_test))) )
 
```

```python
print(rf_run.score(X_train,y_train))
print(rf_run.score(X_test,y_test))
```
